```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import gridspec
```

```python
plt.rcParams["font.family"] = "Arial"
```

```python
my_colors = ["black", "gray", ]
```

```python
d_luad = {
    "SMS": {
        "Metaboverse":1,
        "MNS": 27
        
    },
    "GLYCTK": {
        "Metaboverse":2,
        "MNS": 125
    },
    "UPP1,2": {
        "Metaboverse":4,
        "MNS": 23  
    },
    "XDH": {
        "Metaboverse":11,
        "MNS": 4
        
    }
}
data1 = pd.DataFrame(d_luad).T
data1 = (data1 - 150) * -1
data1 = data1.replace(np.nan, 5)
```

```python
d_mct1 = {
    "LEU5": {
        "Metaboverse":1,
        "MNS": np.nan,
        
    },
    "CAT2": {
        "Metaboverse":1,
        "MNS": np.nan,
        
    },
    "KGD1,2;LPD1": {
        "Metaboverse":2,
        "MNS": 24,
        
    },
    "FXN:NFS1:ISD11:ISCU": {
        "Metaboverse":3,
        "MNS": 243,
        
    },
    "SDH": {
        "Metaboverse":4,
        "MNS": 61,
        
    },
    "COQ": {
        "Metaboverse":4,
        "MNS": 129,
        
    },
    "CTP1": {
        "Metaboverse":6,
        "MNS": np.nan,
        
    },
}
data2 = pd.DataFrame(d_mct1).T
data2 = (data2 - 300) * -1
data2 = data2.replace(np.nan, 5)
```

```python
kw = dict(ha="center", va="center", fontsize=20, color="black")

mm = 1/2.54/10  # centimeters in inches
fig, axs = plt.subplots(2,1, figsize=(44*mm, 180*mm))

ax1 = axs[0]
ax1 = data1.plot.bar(color=my_colors, ax=ax1)
ax1.set_yticks(ticks=[149,139, 124, 99, 24, 5], labels=[1, 10, 25, 50, 125, "not ranked"])
ax1.set_ylabel("Rank")
ax1.legend(loc=9, bbox_to_anchor=(1.2,1))
ax1.set_title("LUAD study")
ax1.text(-0.2, 1.1, "a", transform=ax1.transAxes, **kw)

ax2 = axs[1]
data2.plot.bar(color=my_colors, ax=ax2)
ax2.set_yticks(ticks=[299,275, 250, 200, 50, 5], labels=[1, 25, 50, 100, 250, "not ranked"])
ax2.set_ylabel("Rank")
ax2.legend().set_visible(False)
ax2.set_title("$\it{mct1 \Delta}$ study")
ax2.text(-0.2, 1.1, "b", transform=ax2.transAxes, **kw)

plt.subplots_adjust(left=0.1,
                    bottom=0., 
                    right=2.3, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)

plt.savefig("fig_5-benchmarking.svg", bbox_inches="tight")
```
