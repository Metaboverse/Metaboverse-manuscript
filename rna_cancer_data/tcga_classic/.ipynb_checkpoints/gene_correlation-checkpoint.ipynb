```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
rpkm_data =pd.read_csv(
    'rna_cancer_sample.tsv', 
    sep="\t")
```

```python
rpkm_data = rpkm_data.loc[rpkm_data["Cancer"] == "LUAD"]
```

```python
sns.regplot(
    x=np.log2(rpkm_data.loc[rpkm_data["Gene"] == "ENSG00000102172"]["FPKM"] + 0.1),  # SMS
    y=np.log2(rpkm_data.loc[rpkm_data["Gene"] == "ENSG00000136997"]["FPKM"] + 0.1),  # MYC
    scatter_kws={'alpha': 0.1, 'color': 'black'},
    line_kws={"color": "red"}
)
```

```python
sns.regplot(
    x=np.log2(rpkm_data.loc[rpkm_data["Gene"] == "ENSG00000102172"]["FPKM"] + 0.1),  # SMS
    y=np.log2(rpkm_data.loc[rpkm_data["Gene"] == "ENSG00000171791"]["FPKM"] + 0.1),  # BCL2
    scatter_kws={'alpha': 0.1, 'color': 'black'},
    line_kws={"color": "red"}
)
```

```python
sns.regplot(
    x=np.log2(rpkm_data.loc[rpkm_data["Gene"] == "ENSG00000102172"]["FPKM"] + 0.1),  # SMS
    y=np.log2(rpkm_data.loc[rpkm_data["Gene"] == "ENSG00000153094"]["FPKM"] + 0.1),  # BCL2L11 (BIM)
    scatter_kws={'alpha': 0.1, 'color': 'black'},
    line_kws={"color": "red"}
)
```
