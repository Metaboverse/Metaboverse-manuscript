```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy
```

```python
rpkm_data =pd.read_csv(
    'rna_cancer_sample.tsv', 
    sep="\t")
```

```python
rpkm_data = rpkm_data.loc[rpkm_data["Cancer"] == "LUAD"]
```

```python
sms_corrs = []

for x in list(set(rpkm_data["Gene"])):
    r, p = scipy.stats.pearsonr(
        x=rpkm_data.loc[rpkm_data["Gene"] == "ENSG00000102172"]["FPKM"].values,
        y=rpkm_data.loc[rpkm_data["Gene"] == x]["FPKM"].values)
    sms_corrs.append([x, r, p])
```

```python
sms_corrs.sort(key=lambda x: x[1])
```

```python
myc_corrs = []

for x in rpkm_data["Gene"]:
    r, p = scipy.stats.pearsonr(
        x=rpkm_data.loc[rpkm_data["Gene"] == "ENSG00000136997"]["FPKM"].values,
        y=rpkm_data.loc[rpkm_data["Gene"] == x]["FPKM"].values)

    myc_corrs.append([x, r, p])
```

```python
myc_corrs.sort(key=lambda x: x[1])
```

```python
r, p = scipy.stats.pearsonr(
        x=rpkm_data.loc[rpkm_data["Gene"] == "ENSG00000102172"]["FPKM"].values,
        y=rpkm_data.loc[rpkm_data["Gene"] == "ENSG00000136997"]["FPKM"].values)
print(r, p)

sns.regplot(
    x=np.log2(rpkm_data.loc[rpkm_data["Gene"] == "ENSG00000102172"]["FPKM"] + 0.1),  # SMS
    y=np.log2(rpkm_data.loc[rpkm_data["Gene"] == "ENSG00000136997"]["FPKM"] + 0.1),  # MYC
    scatter_kws={'alpha': 0.1, 'color': 'black'},
    line_kws={"color": "red"}
)
```
